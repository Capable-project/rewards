# Selected Examples from the "*How Can We Reward You?*" paper

## Walking

```python
walking = Score(units=Compliance(10, 20, 40, 61), sessions=Compliance(1, 2, 4, 6), description = "walking", unit_description = "minutes")
walking.plot_units(True)
walking.plot_sessions(True)
```

```python
sessions = [2, 10, 20, 40, 60, 80, 20]
walking.plot_session_score(sessions, show_labels=False)
walking.plot_total_score(sessions)
walking.print_total_score(sessions)
```

## Meditation

### Fruits

```python
fruits = Score(units=Compliance(1, 1, 5, 10, 14), description="fruits")
fruits.plot_units(True)
```

**Note that the code in the next two cells raises exceptions**. This is on purpose to show limitations related to specific settings (e.g., single-session rewards).

`plot_sessions` will fail for a single-session reward as it is not applicable in such a context

```python
fruits.plot_sessions(True)
```

`plot_total_score` and `print_total_score` will fail for a single-session reward -- use `plot_session_sore` instead 

```python
fruits.plot_total_score([2, 10])
```

```python
fruits.plot_session_score([6, 14])
```
